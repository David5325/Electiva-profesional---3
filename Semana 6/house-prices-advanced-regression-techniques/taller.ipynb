#  Dataset: House Prices - Advanced Regression Techniques

---

##  Objetivo del dataset
- **Meta principal**: **Predecir el precio de venta de una casa (`SalePrice`)** en la ciudad de Ames (Iowa, EE.UU.).
- **Cómo se logra**: mediante modelos de **regresión múltiple** u otros algoritmos de Machine Learning, usando como insumo las características de la casa (tamaño, materiales, condiciones, ubicación, etc.).

---

##  Pasos para llegar al objetivo
1. **Comprensión del problema**  
   - Variable dependiente: `SalePrice`.
   - Variables independientes: características estructurales, de ubicación, materiales y extras.

2. **Exploración de los datos (EDA)**  
   - Revisar registros, valores nulos, tipos de datos.
   - Generar estadísticas descriptivas y distribuciones.

3. **Limpieza de datos**  
   - Imputar o eliminar valores faltantes (`NA`).
   - Tratar valores atípicos.
   - Agrupar categorías poco frecuentes.

4. **Codificación de variables categóricas**  
   - Usar *One Hot Encoding* o *Label Encoding*.

5. **Escalamiento / Normalización**  
   - Aplicar sobre variables numéricas cuando el modelo lo requiera.

6. **Selección de variables relevantes**  
   - Usar correlaciones, ANOVA o técnicas de *feature selection*.

7. **Construcción del modelo**  
   - Empezar con **Regresión Lineal Múltiple**.
  

8. **Evaluación del modelo**  
- **RMSE (Root Mean Squared Error):** mide el error promedio, penaliza más los errores grandes.  
- **MAE (Mean Absolute Error):** error promedio absoluto entre lo real y lo predicho.  
- **R² (Coeficiente de determinación):** indica qué tan bien el modelo explica la variabilidad (0 a 1).  
- **Validación cruzada:** divide los datos en partes para entrenar y probar el modelo de forma más robusta.  


9. **Interpretación de resultados**  
   - Identificar qué variables influyen más en el precio de venta.

---

##  Resumen de variables principales
El dataset contiene **79 variables**. Algunos ejemplos clave:

- **Lote**
  - `MSSubClass`: tipo de vivienda (1 piso, 2 pisos, dúplex, PUD).  
  - `MSZoning`: zonificación (residencial, comercial, agrícola).  
  - `LotFrontage`: longitud de fachada.  
  - `LotArea`: área del terreno.  
  - `LotShape`: forma del lote (regular, irregular).  
  - `LandContour`: topografía (plano, pendiente, depresión).  
  - `Neighborhood`: barrio donde se ubica la casa.  

- **Construcción**
  - `BldgType`: tipo de construcción (unifamiliar, dúplex, townhouse).  
  - `HouseStyle`: estilo (1 piso, 2 pisos, split-level).  
  - `OverallQual`: calidad general (1 = muy mala, 10 = excelente).  
  - `OverallCond`: estado general de la vivienda.  
  - `YearBuilt`: año de construcción.  
  - `YearRemodAdd`: año de remodelación.  
  - `RoofStyle` / `RoofMatl`: tipo y material del techo.  
  - `Exterior1st` / `Exterior2nd`: material exterior.  

- **Sótano**
  - `BsmtQual`, `BsmtCond`: calidad y estado del sótano.  
  - `BsmtExposure`: exposición (ventanas, walkout).  
  - `TotalBsmtSF`: área total del sótano.  

- **Habitabilidad**
  - `GrLivArea`: área habitable sobre el nivel del suelo.  
  - `FullBath`, `HalfBath`: baños.  
  - `Bedroom`, `Kitchen`: habitaciones y cocinas.  
  - `TotRmsAbvGrd`: total de cuartos.  

- **Extras**
  - `Fireplaces`, `FireplaceQu`: chimeneas.  
  - `GarageType`, `GarageCars`, `GarageArea`: características del garaje.  
  - `WoodDeckSF`, `OpenPorchSF`, `PoolArea`, `Fence`: áreas exteriores y amenidades.  

- **Venta**
  - `MoSold`, `YrSold`: mes y año de venta.  
  - `SaleType`: tipo de venta (convencional, cash, VA loan).  
  - `SaleCondition`: condiciones de la venta (normal, familiar, parcial, etc.).  

- **Target**
  - `SalePrice`: precio de venta de la vivienda (variable a predecir).

---

##  En resumen
El dataset busca **explicar y predecir el precio de las casas** usando información estructural, de ubicación, calidad de materiales, extras y condiciones de venta.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


tabla_entreno = pd.read_csv("train.csv")
tabla_prueba = pd.read_csv("test.csv")


lista_ids_salida = tabla_prueba["Id"].copy()


print("Entrenamiento: filas, columnas =", tabla_entreno.shape)
print("Prueba: filas, columnas =", tabla_prueba.shape)
print("\nVariables y tipos (train):")
print(tabla_entreno.dtypes.value_counts())

print("\nEstadísticos de SalePrice:")
print(tabla_entreno["SalePrice"].describe())

plt.figure(figsize=(8,4))
sns.histplot(tabla_entreno["SalePrice"], kde=True)
plt.title("Distribución de SalePrice (train)")
plt.show()


n_original_entreno = tabla_entreno.shape[0]
union_temporal = pd.concat([tabla_entreno.drop(columns=["SalePrice"]), tabla_prueba], ignore_index=True)


num_cols_union = union_temporal.select_dtypes(include=[np.number]).columns.tolist()
cat_cols_union = union_temporal.select_dtypes(include=["object"]).columns.tolist()


for c in num_cols_union:
    if c == "Id":
        continue
    med = union_temporal[c].median()
    union_temporal[c].fillna(med, inplace=True)


for c in cat_cols_union:
    union_temporal[c].fillna("Missing", inplace=True)


union_codificada = pd.get_dummies(union_temporal, prefix_sep="__", drop_first=True)

proc_entreno = union_codificada.iloc[:n_original_entreno, :].copy()
proc_prueba = union_codificada.iloc[n_original_entreno:, :].copy()


objetivo_raw = tabla_entreno["SalePrice"].copy()


numeric_after = proc_entreno.select_dtypes(include=[np.number]).columns.tolist()
if "Id" in numeric_after:
    numeric_after.remove("Id")

scaler_local = StandardScaler()
proc_entreno[numeric_after] = scaler_local.fit_transform(proc_entreno[numeric_after])
proc_prueba[numeric_after] = scaler_local.transform(proc_prueba[numeric_after])


y_log = np.log1p(objetivo_raw.values)

selector_peculiar = SelectKBest(score_func=f_regression, k=25)
X_all = proc_entreno.drop(columns=["Id"], errors="ignore")
selector_peculiar.fit(X_all, y_log)
mask_features = selector_peculiar.get_support()
lista_feature_names = X_all.columns[mask_features].tolist()

print("\nFeatures seleccionadas (muestra):", lista_feature_names[:10], " ... (total:", len(lista_feature_names), ")")


X_mat_entreno = X_all[lista_feature_names].copy()
X_mat_prueba = proc_prueba[lista_feature_names].copy()

X_subtrain, X_subval, y_subtrain, y_subval = train_test_split(
    X_mat_entreno, y_log, test_size=0.20, random_state=12345
)


reg_lineal_unico = LinearRegression()
reg_lineal_unico.fit(X_subtrain, y_subtrain)


y_val_log_pred = reg_lineal_unico.predict(X_subval)
y_val_pred_orig = np.expm1(y_val_log_pred)
y_val_true_orig = np.expm1(y_subval)

rmse_val = np.sqrt(mean_squared_error(y_val_true_orig, y_val_pred_orig))
mae_val = mean_absolute_error(y_val_true_orig, y_val_pred_orig)
r2_val = r2_score(y_val_true_orig, y_val_pred_orig)

print("\nEvaluación en validación (escala original):")
print(f"RMSE: {rmse_val:.2f}")
print(f"MAE: {mae_val:.2f}")
print(f"R2: {r2_val:.4f}")


scores_cv = cross_val_score(reg_lineal_unico, X_mat_entreno, y_log, cv=5, scoring="r2")
print("R2 CV (en log-space):", scores_cv)
print("R2 CV promedio:", scores_cv.mean())


# a) Heatmap de correlación entre las features seleccionadas y SalePrice 
df_corr_plot = X_mat_entreno.copy()
df_corr_plot["SalePrice_orig"] = objetivo_raw.values
corr_matrix_local = df_corr_plot.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix_local, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Heatmap correlaciones (features seleccionadas + SalePrice)")
plt.tight_layout()
plt.show()

# b) Scatter real vs predicho 
plt.figure(figsize=(6,6))
sns.scatterplot(x=y_val_true_orig, y=y_val_pred_orig)
plt.plot([y_val_true_orig.min(), y_val_true_orig.max()],
         [y_val_true_orig.min(), y_val_true_orig.max()], linestyle="--", color="black")
plt.xlabel("SalePrice real")
plt.ylabel("SalePrice predicho")
plt.title("Real vs Predicho (validación)")
plt.tight_layout()
plt.show()

# c) Distribución de residuos 
residuales = y_val_true_orig - y_val_pred_orig
plt.figure(figsize=(7,4))
sns.histplot(residuales, kde=True)
plt.title("Distribución de residuales")
plt.show()


y_test_log_pred = reg_lineal_unico.predict(X_mat_prueba)
y_test_pred_orig = np.expm1(y_test_log_pred)

df_salida = pd.DataFrame({"Id": lista_ids_salida, "SalePrice": y_test_pred_orig})
df_salida.to_csv("predicciones.csv", index=False)

print("\nArchivo 'predicciones.csv' generado con éxito ✅")
```
